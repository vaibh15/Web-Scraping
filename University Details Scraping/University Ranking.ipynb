```python
#import required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options


#Adding headless option to avoid browser opening again and again
options = Options()
options.headless = True
options.add_argument('window-size=1920x1080')


#Add target website URL
website='https://www.topuniversities.com/university-rankings/world-university-rankings/2023'

#Add Crome Driver Path here
path=Service('C:/Users/krush/Downloads/chromedriver_win32/chromedriver.exe')
driver=webdriver.Chrome(service=path,options=options)
driver.get(website)
driver.maximize_window()

University_Name=[]
University_Rank=[]
University_Score=[]
University_Location=[]
University_Links=[]
details=[]

#Pagination Code

pagination = driver.find_element(By.XPATH,'//*[@id="alt-style-pagination"]')
page = pagination.find_elements(By.XPATH,'.//li')
last_page = int(page[3].text)         #here I'm taking first 3 pages for now
current_page = 1                      #page[0] is previous button and page[-1] is next button 

   
while current_page <= last_page:
    time.sleep(3)
    container=driver.find_element(By.XPATH,"//*[@id='ranking-data-load']")
    container2=container.find_elements(By.XPATH,".//div[@class='_qs-ranking-data-row normal-row']")
    
#fetch required data here

    for university in container2:
        University_Name.append(university.find_element(By.XPATH,".//a[@class='uni-link']").text)
        University_Rank.append(university.find_element(By.XPATH,'.//div[@class="_univ-rank hide-this-in-mobile-indi "]').text)
        University_Score.append(university.find_element(By.XPATH,".//span[@class='overall-score-span hide-this-in-mobile-indi']").text)
        University_Location.append(university.find_element(By.XPATH,".//div[@class='location ']").text)
        University_Links.append(university.find_element(By.XPATH,".//a[@class='uni-link']").get_attribute('href')) #fetching all university links

    current_page = current_page + 1
    try:
        nxt_btn = pagination.find_element(By.XPATH,'.//a[@class="page-link next"]')
        nxt_btn.click()
    except:
        pass
        
#for loop to fetch data from university links#

for link in University_Links:
    website=link
    path=Service('C:/Users/krush/Downloads/chromedriver_win32/chromedriver.exe')
    driver=webdriver.Chrome(service=path)
    driver.get(website)
    driver.maximize_window()
    time.sleep(5)
    content_list=[]
    try:
        more_link = driver.find_element(By.XPATH,'//*[@id="p2-overview"]/div[3]/div/div/div[1]/p/span/a')
        more_link.click()
        time.sleep(2)
    except:
        pass
    cont=driver.find_elements(By.XPATH,'//*[@id="p2-overview"]/div[3]/div/div/div[2]/p')
    for i in cont:
        content_list.append(i.text)
            

    delim = ""
    res =  [delim.join(content_list)]

    details.extend(res)

driver.quit()
```

```python
df= pd.DataFrame({'University Name':University_Name,'University Rank':University_Rank,'University Score':University_Score,'University Location':University_Location,'Details':details})
df['University Rank'] = df['University Rank'].str.replace('=', '')
df.to_json('universitydetails.json')
df
```
